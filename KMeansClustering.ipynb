# <font color='red'> K-Means Clustering </font> 

K-means clustering is one of the simplest and popular unsupervised machine learning algorithm.  A cluster refers to a collection of data points aggregated together because of certain similarities. Groups similar objects and find the centroids and data objects are grouped together by their centroid distance.

dataset reference link: https://drive.google.com/file/d/1nxr5XYg4JrwB_EtdWdyIwCT4gr85xfzb/view?usp=sharing

importing libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('./KMeansEx.csv') #loading dataset
data #displaying dataset
```

```python
plt.scatter(data['Satisfaction'], data['Loyalty'],c='g')  #plotting satisfaction and loyalty data
plt.xlabel('Satisfaction') #naming x-axis as satisfaction
plt.ylabel('Loyalty') #naming y-axis as loyalty
plt.show() #displaying the graph
```

```python
x=data.copy()  #copying the data
```

```python
kmeans=KMeans(2) #creating two clusters 
kmeans.fit(x) #fitting the data
```

```python
clusters=x.copy() #storing copy of data in clusters variable
clusters['cluster_pred']=kmeans.fit_predict(x)  #predicted values are stored in clusters_pred column
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['cluster_pred'],cmap='rainbow') #plotting the graph with satisfaction and loyalty as x and y axis
plt.xlabel('Satisfaction') #naming x-axis as satisfaction
plt.ylabel('Loyalty') #naming y-axis as loyalty
plt.show() #displaying the graph
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(x) 
x_scaled
```

Elbow method is used for deciding the number of clusters.

```python
s=[]
for i in range(1,30):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    s.append(kmeans.inertia_) 
s
```

```python
plt.plot(range(1,30),s) #displaying the graph of elbow method
plt.xlabel("number of clusters") 
plt.ylabel("s")
plt.show() #displaying the graph
```

```python
kmeans_new=KMeans(4) #creating four clusters
kmeans.fit(x_scaled)
cluster_new=x.copy()
cluster_new['cluster_pred']=kmeans_new.fit_predict(x_scaled)
cluster_new
```

```python
plt.scatter(cluster_new['Satisfaction'], cluster_new['Loyalty'], c= cluster_new['cluster_pred'], cmap='rainbow') #plotting the cluster
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show() #displaying the clustered graph
 
```

* The purple dots represents the people who are less satisfied and less loyal.
* The red dots are people with high loyalty and less satisfaction.
* The yellow dots are the people with high loyalty and high satisfaction.
* The sky blue dots are the people who are in the midst of things.
