# <font color='red'>Support Vector Machine</font>

Support vector machine is used to classify data into two classes by drawing a relationship hyperplane. It is a supervised machine learning algorithm.Support vectors are the data points that lie closest
to the decision surface (or hyperplane).It uses a technique called the kernel trick to transform data.

import libraries

```python
import numpy #consists similar type of data 
import pandas as pd # reads various file formats
import matplotlib.pyplot #represents the data in graphical format
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs #import the dataset


# we create 40 separable points
X, y = make_blobs(n_samples=20, centers=2, random_state=6)

# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel='linear', C=1000)
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='darkblue', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='y')
plt.show()
```

from the dataset blobs, few data samples are scattered and forms two different classes by obtaining a hyperplane.
