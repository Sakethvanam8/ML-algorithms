# <font color='red'>Decision Tree </font>

Decision tree is one of the machine learning algorithms which is used for both classification and regression. This is a tree which splits continuously till its criteria is met. It consists of nodes(test_condition),branch(maps nodes and leaf) and leaf node(output values)

```python
from sklearn.datasets import load_iris #load the dataset iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO 
from IPython.display import Image 
from pydotplus import graph_from_dot_data #install libraries   conda install -c conda-forge python-graphviz
                                                     #conda install pydotplus
import pandas as pd
import numpy as np
import pydotplus as graph_from_dot_data #pydotplus is an interface to Graphviz dot language
from sklearn.tree import export_graphviz #graphviz is a graph drawing software
```

```python
from graphviz import Source
from sklearn import tree
```

```python
iris = load_iris() #load the dataset iris
#load the data components into X and y
X = pd.DataFrame(iris.data, columns=iris.feature_names) 
y = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
X.head() #get first n rows
```

```python
y = pd.get_dummies(y)#it gets converts categorical values into numbers
```

the data is splitted into test set and training set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create Decision Tree classifer object
dt = DecisionTreeClassifier()
# Train Decision Tree Classiferifier()
dt.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = dt.predict(X_test)
```

build decision tree classifier to discover dimensions and type of a flower.

```python
from IPython.display import SVG
graph = Source(tree.export_graphviz(dt, out_file=None
   , feature_names=iris.feature_names, class_names=['0', '1', '2'] 
   , filled = True)) #though decision tree handles categorical data, make use of numerical data.   
display(SVG(graph.pipe(format='svg')))
```
